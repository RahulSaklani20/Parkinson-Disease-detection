```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
x=df.drop(["name","status","NHR",'MDVP:Fhi(Hz)','DFA'],axis=1)
y=df['status']
scaler=StandardScaler()
x=scaler.fit_transform(x)
print(x.shape)
print(y.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=12)
```

KNN

```python
#Checking the best value for k
d = {}
for i in range(2,int(df.shape[0]**0.5)):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    test_pred = knn.predict(x_test)
    ac = accuracy_score(test_pred, y_test)
    d[i] = ac
best_k = max(d, key=d.get)
max_accuracy = d[best_k]
print("Max Accuracy:",max_accuracy, "for k =", best_k)

# Plotting the values
plt.plot(list(d.keys()), list(d.values()))
plt.xlabel('k values')
plt.ylabel('Accuracy')
plt.title('Accuracy for different k values in KNN')
plt.show()
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
train_pred=knn.predict(x_train)
test_pred=knn.predict(x_test)
a1=(accuracy_score(train_pred,y_train)*100)
a2=(accuracy_score(y_test,test_pred)*100)
print("Training Accuracy:",a1)
print("Testing Accuracy:",a2)
```

```python
sns.heatmap(data=confusion_matrix(test_pred,y_test),annot=True)
```

```python
print(classification_report(test_pred,y_test))
```

Support Vector Classifier Kernel=poly

```python
from sklearn.svm import SVC
svc=SVC(kernel='poly')
svc.fit(x_train,y_train)
pred1=svc.predict(x_train)
pred2=svc.predict(x_test)
a3=accuracy_score(pred1,y_train)*100
a4=accuracy_score(pred2,y_test)*100
print("Training Accuracy:",a3)
print("Testing Accuracy:",a4)
```

Support Vector Classifier Kernel=linear

```python
svc=SVC(kernel='linear')
svc.fit(x_train,y_train)
pred3=svc.predict(x_train)
pred4=svc.predict(x_test)
a5,a6=accuracy_score(pred3,y_train)*100,accuracy_score(pred4,y_test)*100
print("Training Accuracy:",a5)
print("Testing Accuracy:",a6)
```

Support Vector Classifier Kernel=rbf

```python
svc=SVC(kernel='rbf')
svc.fit(x_train,y_train)
pred5=svc.predict(x_train)
pred6=svc.predict(x_test)
a7,a8=accuracy_score(pred5,y_train)*100,accuracy_score(pred6,y_test)*100
print("Training Accuracy:",a7)
print("Testing Accuracy:",a8)
```

Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
pred7=dt.predict(x_train)
pred8=dt.predict(x_test)
a9=accuracy_score(pred7,y_train)*100
a10=accuracy_score(pred8,y_test)*100
print("Training Accuracy:",a9)
print("Testing Accuracy:",a10)
```
